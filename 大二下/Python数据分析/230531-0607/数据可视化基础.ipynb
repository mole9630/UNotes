## 1.通过figure()函数创建画布

### 创建空白画布, 返回figure实例

## 2.通过subplot()创建单个子图

```python
nums = np.arange(0, 101)
plt.subplot(2, 2, 1)
plt.plot(nums, nums)
plt.subplot(224) # 参数小于10, 逗号可省略
plt.plot(nums, -nums)
plt.show()
```

## 3.通过subplots()创建多个子图

```python
nums = np.arange(0, 101)
fig, axes = plt.subplots(2, 2) # 分成2*2的矩阵区域, 返回子图数组axes
ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]
ax1.plot(nums, nums)
ax2.plot(nums, -nums)
ax3.plot(nums, nums**2)
ax4.plot(nums, np.log(nums))
plt.show()
```

## 4.添加各类标签

```python
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
# 设置正常显示符号
plt.rcParams['axes.unicode_minus'] = False

nums = np.arange(0, 1.1, 0.01)
plt.title('抛物线')
plt.xlabel('x') # 设置x轴标签
plt.ylabel('y')
plt.xticks([0, 0.5, 1.0]) # 设置x轴刻度
plt.yticks([0, 0.5, 1.0])
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
plt.legend(['y=x^2', 'y=x^3'])
plt.show()
```

## 练习

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.title('正弦余弦')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.xlim(-3, 3)
plt.ylim(-1.2, 1.2)
plt.xticks(np.linspace(-3, 3, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))
plt.plot(x, y_sin, color='g')
plt.plot(x, y_cos, color='r')
plt.legend(['sin(x)', 'cos(x)'])
```

# 二.绘制常见图形

## 1.直方图

```python
data = np.random.standard_normal(1000) # 返回标准正态分布数组
bins = 50
plt.hist(data, bins)
plt.title('直方图')
plt.show()
```

```python
x = np.random.normal(size=200)
plt.hist(x, bins=20, color='m', width=0.2)
plt.show()
```

## 2.柱状图(反映不同类别数据之间分布情况的差异)

```python
data_men = (40, 120, 20, 100, 30, 200)
data_women = (60, 180, 30, 150, 20, 50)
fig, axes = plt.subplots()
index = np.arange(6) # 指定条纹显示的位置
axes.bar(index, data_men, width=0.4, color='#c7edcc', label='men')
axes.bar(index+0.4, data_women, width=0.4, color='#f7edcc', label='woman') # 坐标轴向右平移0.4个单位
axes.set_xticks(index+0.4/2) # 设置x轴刻度
axes.set_xticklabels(('出租车', '地铁', '步行', '公交车', '自行车', '驾车')) # 刻度标签
axes.legend()
plt.show()
```

```python
data_men = (40, 120, 20, 100, 30, 200)
data_women = (60, 180, 30, 150, 20, 50)
width=0.4
fig, axes = plt.subplots()
index = np.arange(6) # 指定条纹显示的位置
axes.barh(index, data_men, width, color='#c7edcc', label='men')
axes.barh(index+0.4, data_women, width, color='#f7edcc', label='woman') # 坐标轴向下平移0.4个单位
axes.set_yticks(index+0.4/2) # 设置x轴刻度
axes.set_yticklabels(('出租车', '地铁', '步行', '公交车', '自行车', '驾车')) # 刻度标签
axes.legend()
plt.show()
```
