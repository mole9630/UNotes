### 异常值处理

#### 3σ原则

```python
def three_sigma(ser):
    """
    :param ser: 被检测的数据，接收DataFrame的一列数据
    :return: 异常值及其对应的行索引
    """
    mean_data = ser.mean() # 计算平均值
    std_data = ser.std() # 计算标准差
    #小于μ-3σ或大于μ+3σ的数值均为异常值
    rule = (mean_data-3*std_data>ser) | (mean_data+3*std_data<ser)
    index = np.arange(ser.shape[0])[rule] # 返回异常值的行索引
    outliers = ser.iloc[index]  # 获取异常值
    return outliers
```

```python
three_sigma(data['房价中位数'])
```

# 案例-运动员信息的分组与聚合

## 1.读取数据

```python
data = pd.read_csv(r'./data/运动员信息表.csv', encoding='gbk')
data
```

## 2.筛选篮球运动员

```python
data_group = data.groupby('项目')
basketball = data_group.get_group('篮球')
basketball
```

## 3.统计篮球运动员的平均年龄、身高、体重

```python
basketball_group = basketball.groupby('性别')
basketball_group.mean()
```

## 4.统计男篮球运动员的年龄、身高、体重的极差值

```python
def func(x): # 求极差值
    return max(x) - min(x)

basketball_group_man = basketball_group.get_group('男')
basketball_group_man.agg({'年龄（岁）': func, '身高(cm)': func, '体重(kg)': func})
```

## 5.统计篮球运动员的体质指数(BMI)

```python
BMI = DataFrame({
    '体重指数':['过轻','正常','过重','肥胖','极度肥胖'],
    '男性':['低于20','20-25','25-30','30-35','高于35'],
    '女性':['低于19','19-24','24-29','29-34','高于34']
})
BMI
```

```python
basketball['体质指数BMI'] = (basketball['体重(kg)']) / (((basketball['身高(cm)']/100)**2))
basketball
```

## 6.给运动员BMI添加标签

```python
bins = [0, 20, 25, 30, 35, 50]
labels = ['过轻', '正常', '过胖', '肥胖', '极度肥胖']
basketball['BMI标签'] = pd.cut(basketball['体质指数BMI'], bins, labels=labels)
basketball
```

```python
basketball[basketball['BMI标签'] != '正常']
```
