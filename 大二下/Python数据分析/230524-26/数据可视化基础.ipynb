# 一.matplotlib-绘制图表

## 1.通过figure()函数创建画布

```python
data1 = np.arange(100, 200)
plt.plot(data1)
plt.show()
```

### 创建空白画布, 返回figure实例

```python
figure = plt.figure() # 如不想在默认画布绘制图像, 可以创建一个新的空白画布
data2 = np.arange(0, 10)
plt.figure(facecolor='#c7edcc', figsize=(5, 5)) # 创建自定义画布
plt.plot(data2)
plt.show()
```

```python
plt.plot(data2, data2**2)
plt.show()
```
