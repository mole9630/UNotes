## 3.用箱型图检查数据是否有异常值

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
beijing_data.boxplot(column=['行政面积（K㎡）', '户籍人口（万人）', '男性', '女性', 'GDP（亿元）', '常住人口（万人）'])
```

```python
def box_outliers(ser):
    # 对待检测的数据集进行排序
    new_ser = ser.sort_values()
    # 判断数据的总数量是奇数还是偶数
    if new_ser.count() % 2 == 0:
        # 计算Q3、Q1、IQR
        Q3 = new_ser[int(len(new_ser) / 2):].median()
        Q1 = new_ser[:int(len(new_ser) / 2)].median()
    elif new_ser.count() % 2 != 0:
        Q3 = new_ser[int((len(new_ser)-1) / 2):].median()
        Q1 = new_ser[:int((len(new_ser)-1) / 2)].median()
    IQR = round(Q3 - Q1, 1)
    rule = (round(Q3+1.5*IQR, 1) < ser)|(round(Q1-1.5*IQR, 1) > ser)
    index = np.arange(ser.shape[0])[rule]
    # 获取异常值及其索引
    outliers = ser.iloc[index]
    return outliers
```

```python
box_outliers(beijing_data['GDP（亿元）']) # 经济强区, 不作处理
```

# 二.中国运动员信息分析

## 1.读取运动员数据并采用纵向堆叠方式合并数据

```python
athlete1 = pd.read_csv(r'./data/运动员信息采集01.csv', encoding='gbk')
athlete2 = pd.read_excel(r'./data/运动员信息采集02.xlsx', encoding='gbk')
athlete = pd.concat([athlete1,  athlete2], ignore_index=True) # 纵向堆叠, 并重新索引
athlete.head()
```

## 2.用布尔选取, 选择国籍为中国的运动员

```python
athlete_china = athlete[athlete['国籍'] == '中国']
athlete_china.head()
```

## 3.检查重复值并删除

```python
athlete_china.duplicated()
```

```python
athlete_china[athlete_china.duplicated().values == True]
```

```python
athlete_china = athlete_china.drop_duplicates() # 删除重复值
athlete_china[athlete_china.duplicated().values == True]
```

## 4.筛选出篮球运动员

```python
athlete_china_basketball = athlete_china[athlete_china['项目']=='篮球']
athlete_china_basketball.head()
```

```python
athlete_china_basketball.rename(columns={'身高':'身高(cm)', '体重':'体重(kg)'}, inplace=True)
athlete_china_basketball.head()
```

## 5.将运动员按性别分类

```python
woman = athlete_china_basketball[athlete_china_basketball['性别']=='女']
woman.head()
```

```python
woman.isnull().sum()
```

```python
woman_height_mean = round(woman['身高(cm)'].mean(), 1)
woman_height_mean
```

```python
woman_weight_mean = round(woman['体重(kg)'].mean(), 1)
woman_weight_mean
```

```python
woman = woman.fillna({'身高(cm)': woman_height_mean, '体重(kg)': woman_weight_mean})
woman.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
woman.boxplot(column=['身高(cm)', '体重(kg)'])
```

```python
box_outliers(woman['身高(cm)']) # 实际无异常, 不作处理
```
