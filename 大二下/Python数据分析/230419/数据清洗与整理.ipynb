## 2.缺失值删除

```python
dict = {"类别":['小说', '散文随笔', '青春文学', '传记'],
           "书名":[np.nan, '《皮囊》', '《旅程结束时》', '《老舍自传》'],
           "作者":["老舍", None, "张其鑫", "老舍"]
}
df = DataFrame(dict)
df
```

```python
df.dropna() #按照缺失值的行删除
```

```python
df.dropna(axis=0, how='any') #axix=0缺省值, 按行删除;how='any'缺省值, 表示只要有缺失值就删除
```

```python
df.dropna(how='all') #值为'all'表示全部为空才删除
```

```python
df.dropna(axis=1) #按照缺失值的列删除
```

### 练习2

```python
penguins = pd.read_csv(r'.\data\penguins.csv',index_col=0,encoding='utf-8') #index_col=0用第一列数据作为行所以
penguins.head(10)
```

```python
col_bill_length_mm = round(np.mean(penguins['bill_length_mm']), 1)
col_bill_depth_mm = round(np.mean(penguins['bill_depth_mm']), 1)
col_flipper_length_mm = round(np.mean(penguins['flipper_length_mm']), 1)
col_body_mass_g = round(np.mean(penguins['body_mass_g']), 1)
col_sex = penguins['sex'].mode().values[0]
```

```python
penguins = penguins.fillna({'bill_length_mm':col_bill_length_mm,'bill_depth_mm':col_bill_depth_mm, 'flipper_length_mm':col_flipper_length_mm, 'body_mass_g':col_body_mass_g, 'sex':col_sex})
penguins.head(10)
```

### 2).通过fillna()的method参数填充

```python
df.fillna(method='ffill')
```
