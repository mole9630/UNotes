```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
```

```python
dict = {
    "sex":{"小张":"女","小李":"男","小王":"男","小马":"女"},
    "city":{"小张":"陕西","小李":"云南","小王":"广西"}
}
df = DataFrame(dict)
df
```

```python
df['score'] = [60, 70, 80, 90] # 在列尾插入score列
df.insert(1, 'class', [1, 2, 3, 4]) # 在'sex'后添加'class'列
df
```

```python
df.loc['小明'] = ['男', 8, '安徽', 100]
df
```

```python
df.iloc[3] = ['女', 4, '上海', 90] # 修改'小马'的'city'列
df['city']['小马'] = '北京'
df
```

```python
df.drop(['小李'], inplace=True) # 删除'小李'行
df
```
