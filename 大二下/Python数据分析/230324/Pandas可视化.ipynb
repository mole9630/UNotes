## 1.常规绘图方法

```python
plt.plot([1, 2, 3, 4, 5], [1, 6, 5, 19, 10])
```

```python
df = DataFrame({
    'N1': np.random.normal(size=30),
    'N2': np.random.gamma(1, size=30),
    'N3': np.random.poisson(size=30)
})
df
```

```python
df.plot()
```

```python
df.cumsum().plot() # 求累加和可视图
```

```python
dict={
    'name':['陈五','王四','林一','马三'],
    'sex':['female','male','male','male'],
    'year':[2000,2001,2019,2002],
    'city':['武汉','云南','广州','上海']
}
df = DataFrame(dict)
df
```

```python
df['sex'].value_counts()
```

```python
df['sex'].value_counts().plot(kind='bar', color='#c7edcc')
```

```python
df['sex'].value_counts().plot(kind='barh', color='#c7edcc') # 条形图
```

## 2.直方图和密度图

```python
s1 = Series(np.random.normal(size=80)) # 直方图
s1.plot(kind='hist', bins=20, color='#c7edcc')
```

```python
s1.plot(kind='kde', color='#c7edcc') # 密度图
```
